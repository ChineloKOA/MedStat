```python
import statistics_functions

from helper.aws_helper import S3Connection
import statistics_functions
import pandas as pd
```

```python
# instantiate and read dataset
conn = S3Connection()
medsafe_dataset = conn.read_file("medsafe-docs/physicians_migration_additional_cleaned_data.csv",
                                 "medsafe-docs")
```

```python
# provide the df for the professional cadre column
professional_cadre_df = statistics_functions.provide_df(medsafe_dataset, "9. What is your current professional cadre?", "Profession","Count")
# provide the df for the gender column
gender_cadre_df = statistics_functions.provide_df(medsafe_dataset, "2. Your sex", "Gender","Count")
# provide the df for the religion column
Religion_cadre_df = statistics_functions.provide_df(medsafe_dataset, "5. Your religion. (Please, specify the religion if your response is 'Other')", "Religion","Count")
# provide the df for the Marital status column
Marital_cadre_df = statistics_functions.provide_df(medsafe_dataset, "3. Marital status", "Marital status","Count")
dataframes_list = [professional_cadre_df, gender_cadre_df, Religion_cadre_df, Marital_cadre_df]
titles_list = ["Professional cadre", "Gender distribution", "Religious Distribution", "Marital status"]
statistics_functions.multi_h_bar(dataframes_list, titles_list)
```

#### Above is a basic distribution of medical profession cadre, gender & religion distribution and maritial status of those surveyed

```python
Satisfaction_distribution_df = statistics_functions.provide_df(medsafe_dataset, "15. How satisfied are you with your current professional practice in Nigeria?","Level of Satisfaction", "Total Count")
dataframe = Satisfaction_distribution_df
title = "Satisfaction level of medical professionals practicing in Nigeria"
statistics_functions.single_v_bar(dataframe, title)
```

```python
Unsatisfied_table = statistics_functions.distributions(medsafe_dataset, "15. How satisfied are you with your current professional practice in Nigeria?", "9. What is your current professional cadre?", "Unsatisfied")
Very_Unsatisfied_table = statistics_functions.distributions(medsafe_dataset, "15. How satisfied are you with your current professional practice in Nigeria?", "9. What is your current professional cadre?", "Very unsatisfied")
Satisfied_table = statistics_functions.distributions(medsafe_dataset, "15. How satisfied are you with your current professional practice in Nigeria?", "9. What is your current professional cadre?", "Satisfied")
Very_Satisfied_table = statistics_functions.distributions(medsafe_dataset, "15. How satisfied are you with your current professional practice in Nigeria?", "9. What is your current professional cadre?", "Very satisfied")
dataframes_list = [Very_Unsatisfied_table, Unsatisfied_table, Very_Satisfied_table, Satisfied_table]
titles_list = ['% of Professions that are Very unsatisfied', '% of Professions that are Unsatisfied', '% of Professions that are Very satisfied', '% of Professions that are Satisfied']
statistics_functions.multi_pie_chart(dataframes_list, titles_list)
```
